```python
df = df.drop('id',axis=1)
```

```python
df.info() # all the features are in object so change them into float
```

```python
df = df.astype('float64')
df.info()
```

```python
# Create correlation matrix
corr_matrix = df.corr().abs()

# Select upper triangle of correlation matrix
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(np.bool))

# Find index of feature columns with correlation greater than 0.95
to_drop = [column for column in upper.columns if any(upper[column] > 0.75)]
```

Finally, top twenty features werecalculated utilizing Kbest algorithm that takes a scoringfunction as a parameter, which must be relevant to a pair(X, y). The score function must return an array of scores,one for each feature and then simply holds thefirst â€˜kâ€™ features of â€˜Xâ€™ with the highest scores 


```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2

y = df['class']
X = df.drop('class',axis=1)
X = X.abs()
X

chi2_features = SelectKBest(chi2,k = 20)
new_df = chi2_features.fit_transform(X, y)
```

```python
new_df = pd.DataFrame(new_df)
new_df.head()
```

```python
# there the reduced features are from the kbest algo we got the best 20 features
columns = ['numPulses','GNE_NSR_SEO','mean_MFCC_1st_coef','mean_MFCC_6th_coef','std_MFCC_8th_coef','tqwt_energy_dec_27','tqwt_energy_dec_31','tqwt_energy_dec_35','tqwt_entropy_shannon_dec_32','tqwt_TKEO_mean_dec_21','tqwt_medianValue_dec_25','tqwt_medianValue_dec_26','tqwt_medianValue_dec_34','tqwt_meanValue_dec_9','tqwt_meanValue_dec_11','tqwt_meanValue_dec_21','tqwt_meanValue_dec_26','tqwt_skewnessValue_dec_3','tqwt_skewnessValue_dec_21','tqwt_skewnessValue_dec_28','class']
reduced_df = df[columns]
reduced_df.head()
```
